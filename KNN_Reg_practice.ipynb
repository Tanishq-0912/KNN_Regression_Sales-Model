```python
# Normalization of X

norm_scaler = MinMaxScaler()
X_nor = norm_scaler.fit_transform(X)
X_nor_df = pd.DataFrame(X_nor,columns =X.columns)
X_nor_df
```

```python
# Train test split

X_train,X_test,y_train,y_test = train_test_split(X_nor_df,y,test_size=0.15,random_state=42)
```

```python
X_train.shape
```

```python
# Train Model

knn_reg = KNeighborsRegressor(n_neighbors=10,p=2)
knn_reg.fit(X_train,y_train)
```

```python
y_pred= knn_reg.predict(X_test)
y_pred
```

# Evaluation

```python
# Mean squared error
mse = mean_squared_error(y_test,y_pred)
print(mse)
```

```python
#Root mean squared error
rmse = np.sqrt(mse)
rmse
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
accuracy =r2_score(y_test,y_pred)
accuracy
```
